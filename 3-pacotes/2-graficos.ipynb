# Gráficos

## Básico

Existem diversas formas de criar gráficos com Julia. Podemos fazer isso através de funções do R e Python dentro do Julia, através do PyPlot, etc; existe uma infinitude de pacotes!

Para este tutorial, no entanto, iremos utilizar o pacote `Plots.jl`:

```julia
#Pkg.add("Plots")

using Plots
```

A vantagem de utilizarmos o pacote `Plots.jl` é que podemos usar diferentes *backends* sem alterar os comandos que nós utilizamos. Em outras palavras, diferentes algoritmos/programas/pacotes são utilizados sem que isso afete a experiência do usuário de maneira significativa. Para este tutorial, iremos utilizar os backends `gr()` e `plotlyjs()`.

Antes disso, no entanto, iremos gerar dados!

```julia
x = -3:0.1:3
f(x)=x^2

y=f.(x)
```

Agora, carreguemos o backend GR:

```julia
gr()
```

```julia
plot(x,y, label="line")
scatter!(x,y,label="points")
```

O sinal de exclamação em `scatter!()` significa que iremos **adicionar** pontos de dispersão ao gráfico de linha que já foi definido anteriormente.

E se quisermos utilizar o backend `plotlyjs()`? Veja como os gráficos são diferentes, mas a sintaxe não muda nem um pouco:

```julia
plotlyjs()
```

```julia
plot(x,y, label="line")
scatter!(x,y,label="points")
```

## Incrementando nossos gráficos

A sintaxe para adicionarmos títulos e legendas para nossos gráficos é bastante simples.

Para fins científicos, iremos analisar a relação entre a temperatura global e o número aproximado de piratadas entre 1860 e 2000:

```julia
temperaturas=[14.4,14.5,14.8,15.2,15.5,15.8]
piratas=[45000,20000,15000,5000,400,13]
```

```julia
# Primeiro, plotamos o gráfico

plot(piratas, temperaturas,legend=false)
scatter!(piratas, temperaturas, legend=false)

# Isso inverte o eixo x do grafico para visualizarmos melhor a mudanca de temperaturas

xflip!()

# Adicionando titulo e legendas aos eixos

xlabel!("Número de piratas (aproximado)")
ylabel!("Temperatura (C)")
title!("Influência da população pirata na temperatura global")
```

Para nosso último exemplo, iremos criar criar subgráficos. Para isso, só precisamos definir cada gráfico como um objeto e "chamá-los" num comando `plot()` definindo layout e quais gráficos irão ser chamados:

```julia
p1=plot(x,x)
p2=plot(x, x.^2)
p3=plot(x, x.^3)
p4=plot(x, x.^4)

plot(p1,p2,p3,p4,layout=(2,2),legend=false)
```
